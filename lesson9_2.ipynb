```python
#自訂function,參數
def say_hello():
    print("Hello")
```

```python
#呼叫function,引數值
say_hello()
```

```python
def say_hello(name):
    print(f'Hello!{name}')
```

```python
#引數值的呼叫
say_hello("徐國堂")
```

```python
#引數名稱的呼叫
say_hello(name="徐國堂")
```

```python
def menu(wine, entree, dessert):
    print(f'前酒:{wine}')
    print(f'主餐:{entree}')
    print(f'甜點:{dessert}')
```

```python
#引數值呼叫,必需依照參數的順序
menu('白酒','牛排','蛋糕')
```

```python
#引數名稱的呼叫,不需要依順序
menu(entree="牛排",dessert='蛋糕',wine='白酒')
```

```python
#混合呼叫
#前面使用引數值
#後引數名稱呼叫
menu('白酒',dessert='蛋糕',entree='牛排')
```

```python
#參數可以有type hint

def menu(wine:str, entree:str, dessert:str):
    print(f'前酒:{wine}')
    print(f'主餐:{entree}')
    print(f'甜點:{dessert}')
```

```python
menu('白酒',dessert='蛋糕',entree='牛排')
```

```python
#function傳出的值
def menu(wine:str, entree:str, dessert:str)->dict[str,str]:    
    return {'前酒':wine,'主餐':entree,'甜點':dessert}
```

```python
my_menu = menu('白酒',dessert='蛋糕',entree='牛排')
my_menu
```

```python
#參數的default value
def menu(wine:str, entree:str, dessert:str='蛋糕')->dict[str,str]:    
    return {'前酒':wine,'主餐':entree,'甜點':dessert}
```

```python
print(menu('白酒','牛排'))
print(menu('白酒','牛排',dessert='水果'))
```

```python
def menu(wine:str, entree:str='牛排', dessert:str='蛋糕')->dict[str,str]:    
    return {'前酒':wine,'主餐':entree,'甜點':dessert}
```

```python
print(menu('白酒'))
print(menu('白酒',dessert='水果'))
print(menu('白酒',dessert='水果',entree='雞排'))
```

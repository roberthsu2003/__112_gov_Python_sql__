```python
import psycopg2
import os
from psycopg2 import connect
from dotenv import load_dotenv

#初始化環境變數
load_dotenv()
```

## 建立資料表student

```python
#使用環境變數POSTGRE_PASSWORD,取得密碼,並建立connect(連線)的實體
conn:connect = psycopg2.connect(os.environ['POSTGRE_PASSWORD'])

# 建立cursor實體,準備執行SQL
cur = conn.cursor()

sql = '''
CREATE TABLE IF NOT EXISTS student(
	id SERIAL PRIMARY KEY,
	name VARCHAR(20),
	chinese SMALLINT,
	english SMALLINT,
	math SMALLINT
);
'''

#執行sql
cur.execute(sql)

#確認SQL可以執行
conn.commit()

#關閉cursor
cur.close()

#關閉連線
conn.close()
```

## 新增資料

```python
#使用環境變數POSTGRE_PASSWORD,取得密碼,並建立connect(連線)的實體
conn:connect = psycopg2.connect(os.environ['POSTGRE_PASSWORD'])

# 建立cursor實體,準備執行SQL
cur = conn.cursor()

sql = '''
INSERT INTO student(name, chinese, english, math)
VALUES(%(name)s,%(chinese)s,%(english)s,%(math)s);
'''

#執行sql
cur.execute(sql,{'name':'張國堂','chinese':89, 'english':90, 'math':78})

#確認SQL可以執行
conn.commit()

#關閉cursor
cur.close()

#關閉連線
conn.close()
```

## 新增多筆資料

```python
#使用環境變數POSTGRE_PASSWORD,取得密碼,並建立connect(連線)的實體
conn:connect = psycopg2.connect(os.environ['POSTGRE_PASSWORD'])

# 建立cursor實體,準備執行SQL
cur = conn.cursor()

sql = '''
INSERT INTO student(name, chinese, english, math)
VALUES(%(name)s,%(chinese)s,%(english)s,%(math)s);
'''

#執行sql
data:list[dict] = [
    {'name':'王國堂','chinese':89, 'english':75, 'math':92},
    {'name':'陳國堂','chinese':65, 'english':93, 'math':59},
    {'name':'邱國堂','chinese':71, 'english':64, 'math':92}
]
for item in data:
    cur.execute(sql,item)

#確認SQL可以執行
conn.commit()

#關閉cursor
cur.close()

#關閉連線
conn.close()
```

## 顯示資料表

```python
#使用環境變數POSTGRE_PASSWORD,取得密碼,並建立connect(連線)的實體
conn:connect = psycopg2.connect(os.environ['POSTGRE_PASSWORD'])

# 建立cursor實體,準備執行SQL
cur = conn.cursor()

sql = '''
SELECT *
FROM student
'''

#執行sql
cur.execute(sql)

#取得資料
print(cur.fetchall())

#確認SQL可以執行
conn.commit()

#關閉cursor
cur.close()

#關閉連線
conn.close()
```
